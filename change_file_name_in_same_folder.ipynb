### Change File Name in Same Folder

```python
import os

def rename_images(folder_path, old_base_name, new_base_name, extension, fil_count):

    try:
        files = os.listdir(folder_path)
    except FileNotFoundError:
        print("경로를 찾을 수 없습니다. 폴더 경로를 확인해 주세요.")
        return

    count = 0
    for i in range(fil_count):  # 0부터 41까지 반복
        old_name = f"{old_base_name} ({i})" + extension
        new_name = f"{new_base_name} {i:02d}" + extension
        
        # 파일이 실제로 존재하는지 확인 후 변경
        if old_name in files:
            old_path = os.path.join(folder_path, old_name)
            new_path = os.path.join(folder_path, new_name)
            
            os.rename(old_path, new_path)
            print(f"변경 완료: {old_name} -> {new_name}")
            count += 1
        else:
            print(f"찾을 수 없음: {old_name}")

    print(f"--- 총 {count}개의 파일 이름이 변경되었습니다. ---")
```

```python
folder_path = './file_name'
old_base_name = "image"
new_base_name = "중동 귀족 가문의 관계"
extension = ".png"
fil_count = 42
rename_images(folder_path, old_base_name, new_base_name, extension, fil_count)
```
