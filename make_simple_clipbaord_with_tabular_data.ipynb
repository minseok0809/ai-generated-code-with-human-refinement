### Make Simple Clipboard With Tabular Data

#### Example - 러시아 행정구역 (나무위키)

```python
text = """번호
이름
면적(㎢)
인구(명)
수도
1
다게스탄 공화국
Республика Дагестан
50,270
3,063,885
마하치칼라
Махачкала
2
인구셰티야 공화국
Республика Ингушетия
3,628
488,043
마가스
Магас
3
카바르디노-발카리야 공화국
Кабардино-Балкарская Республика
12,470
865,828
날치크
Нальчик
4
카라차예보-체르케시야 공화국
Карачаево-Черкесская Республика
14,277
466,305
체르케스크
Черкесск
5
북오세티야 공화국 (알라니야)
Республика Северная Осетия - Алания
7,987
701,765
블라디캅카스
Владикавказ
6
스타브로폴 지방
Ставропрльский край
66,160
2,800,674
스타브로폴
Ставрополь
7
체첸 공화국
Чеченская Республика
15,647
1,436,981
그로즈니
Грозный
"""



# 1. 줄바꿈 기준으로 텍스트 분리 및 공백 제거
lines = [line.strip() for line in text.strip().split('\n') if line.strip()]

# 2. 헤더 제외, 본문 데이터 시작 위치 찾기 (숫자 '1'이 나오는 지점)
start_index = lines.index('1')
data_body = lines[start_index:]

# 3. 규칙 찾기: 
# 각 레코드는 [번호, 이름(한), 이름(러), 면적, 인구, 수도(한), 수도(러)] 순서로 7개씩 반복됨
names = []
step = 7

for i in range(0, len(data_body), step):
    # '합계'가 나오면 중단
    if data_body[i] == '합계':
        break
    
    # 이름(한글)과 이름(러시아어)을 합쳐서 저장
    name_kor = data_body[i + 1]
    name_rus = data_body[i + 2]
    full_name = f"{name_kor}({name_rus})"
    names.append(full_name)

# 결과 출력
results = []
for idx, name in enumerate(names, 1):
    results.append(name)
    # print(f"[{idx}] {name}")
    # print("-" * 20)
```
