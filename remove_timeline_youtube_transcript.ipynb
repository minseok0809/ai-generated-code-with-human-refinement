### Remove Timeline Youtube Transcript

Source:
[Man Drives 100+ Miles For Birth Of Child That Woman Says Isn't His (Full Episode) | Paternity Court](https://www.youtube.com/watch?v=13i0ViDkH8E)

```python
import re

# 원본 텍스트
text = """11:0711 minutes,   7 seconds  MILLER: I have told him a couple of times that if he would just sign her over to me then he wouldn't ha
0:5757 seconds  I drive 160 miles, every other weekend just to pick up my daughter and drop her off. Over... That's about 10,000 miles.
1:051 minute,   5 seconds  RASMUSSEN: That's wear and tear on my car. She doesn't have a driver's license, she doesn't """

# 정규표현식 패턴 설명:
# \d+:\d+          -> '11:07'과 같은 숫자:숫자 형태
# .+?              -> 최소한의 문자를 포함 (minute, seconds 등)
# seconds\s* -> 'seconds' 단어와 그 뒤의 공백까지 포함
pattern = r"\d+:\d+.+?seconds\s*"

# re.sub를 사용하여 해당 패턴을 빈 문자열로 대체
cleaned_text = re.sub(pattern, "", text)

print(cleaned_text)
```
